```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://m.imdb.com/list/ls063915608/')
```

```python
page
```

```python
A=BeautifulSoup(page.content)
```

```python
titles=[]
cnt=0
for i in A.find_all('span',class_="h4"):
    if cnt%2!=0:
        titles.append(i.text)
    cnt+=1
titles
```

```python
ratings=[]
for j in A.find_all('span',class_="imdb-rating"):
    ratings.append(j.text.replace("[]" , " "))
ratings
```

```python
years=[]
cnt=0
for k in A.find_all('span',class_="nowrap"):
    if cnt%2==0:
        years.append(k.text)
    cnt+=1    
years
```

```python
import pandas as pd
df= pd.DataFrame({'Name of movie':titles,'Rating':ratings,'Year of release':years})
df
```
