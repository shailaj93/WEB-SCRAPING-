```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi")
page
```

```python
A=BeautifulSoup(page.content)
A
```

```python
name=A.find('div',class_="rankings-block__banner--name")
names=[]
cnt=0
for i in A.find_all('td',class_='table-body__cell name'):
    if cnt<9:
        names.append(i.text.strip())
    cnt+=1
names.insert(0,name.text)
names
```

```python
team=A.find('div',class_="rankings-block__banner--nationality")
teams=[]
cnt=0
for j in A.find_all('span',class_='table-body__logo-text'):
    if cnt<9:
        teams.append(j.text)
    cnt+=1
teams.insert(0,team.text.split()[0])
teams
```

```python
rating=A.find('div',class_="rankings-block__banner--rating")
ratings=[]
cnt=0
for j in A.find_all('td',class_='table-body__cell u-text-right rating'): 
    if cnt<9:
        ratings.append(j.text)
    cnt+=1
ratings.insert(0,rating.text)
ratings
```

```python
print(len(names),len(teams),len(ratings))
```

```python
import pandas as pd
df= pd.DataFrame({'Name of Team':names,'Team':teams,'Ratings':ratings})
df
```
