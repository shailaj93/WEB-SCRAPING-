```python
A=BeautifulSoup(page.content)
A
```

```python
House_title=[]
for i in A.find_all('h2',class_="heading-6 font-semi-bold nb__25Cl7"):
        House_title.append(i.find('span').text)
House_title
```

```python
location=[]
for j in A.find_all('div',class_='nb__nXU01'):
        location.append(j.find('div').text)
location
```

```python
area=[]
for i in A.find_all('div',class_='nb__FfHqA'):
    area.append(i.text)
area
```

```python
EMI=[]
cnt=0
for l in A.find_all('div',class_='font-semi-bold heading-6',id="roomType"):
        EMI.append(l.text)
EMI
```

```python
price=[]
for m in A.find_all('div',class_="nb__4L90a", id="minDeposit", itemprop="valueReference"):
    price.append(m.find("div").text)
price
```
